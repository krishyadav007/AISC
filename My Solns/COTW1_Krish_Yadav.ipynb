# Code

## Importing libraries and environment setup

```
import datetime as dt
import calendar as cl
import time

class bcolors:
    HEADER = '\033[95m'
    OKBLUE = '\033[94m'
    OKCYAN = '\033[96m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'

def VERBOSE_FUN(para, color = bcolors.HEADER): # Verbose function to print verbose
  if VERBOSE : print( color + "VERBOSE : " + str(para) + bcolors.ENDC)

VERBOSE = True #Set verbose true if you want to keep seeing the process output

VERBOSE_FUN('VERBOSE IS ACTIVATED',bcolors.WARNING)
print( bcolors.OKGREEN + "ENVIRONMENT SETUP SUCCESSFULLY COMPLETED" + bcolors.ENDC)
```

## Taking user input

```
print('Please enter the date of birth below. Example :19/10/1995 5:00 PM')
inpt = input('Enter date here in [dd/mm/yyyy hh:mm AM/PM] format : ')

print()
print('Your date of birth has been recived as ' + inpt)

VERBOSE_FUN('The input is ' + inpt)
```

## Processing

```
VERBOSE_FUN('Entered proccessing')
inpt = inpt.split('/') # ['19','10','1995 5:00 PM']
VERBOSE_FUN('Step 1.1 | The input is ' + str(inpt))

inpt.extend(inpt[2].split(' ')) # Need to extend the list or else it will for a nested list ex : ['19','10',['1995', '5:00', 'PM']]
inpt.pop(2) # Poping the unnecsary orignial entry. Now inpt = ['19','10','1995', '5:00', 'PM']
VERBOSE_FUN('Step 1.2 | The input is ' + str(inpt))

inpt.extend(inpt[3].split(':'))
inpt.pop(3) 
VERBOSE_FUN('Step 1.3 | The input is ' + str(inpt))

inpt.append(inpt[3])
inpt.pop(3) #Some rearranging. ['19', '10', '1995', '5', '00', 'PM']
VERBOSE_FUN('Step 1.4 | The input is ' + str(inpt))

cur = dt.datetime.now() #Current date time

for i in range(5):
  inpt[i] = int(inpt[i]) #Converting to integer
VERBOSE_FUN('Step 2.1 | The input is ' + str(inpt))

if inpt[4] == 'PM':
  inpt[4] += 12 # Converting to 24-Hour clock [19, 10, 1995, 17, 00, 'PM']
VERBOSE_FUN('Step 2.2 | The input is ' + str(inpt))

dob = dt.datetime(inpt[2], inpt[1], inpt[0], inpt[3], inpt[4]) #(year, month, day, hour=0, minute=0); Further , second=0, microsecond=0, tzinfo=None, *, fold=0)
VERBOSE_FUN('Step 3.1 | The D.O.B object is ' + str(dob))

diff = cur - dob #Diffrence
VERBOSE_FUN('Step 3.2 | The diffrence is ' + str(diff))

print('You have been on this planet for ' + bcolors.FAIL + str(diff.days) + bcolors.ENDC + ' Days or ' + bcolors.FAIL + str((diff.days)//30) + bcolors.ENDC + ' months')
```

## Advance output

```
print('Advance output ‚è∞')
print('You have lived for :' + str(diff))

print('You have slept üò¥ for :' + str((diff.days)* 8/24) + 'days' ) #Considering that a human sleeps 8 hours on average

if input('Would you like to see calendar üìÖ of each year you have lived (Y/N)').lower()[0] == 'y':
  print('Calendar of each year you have lived :')
  for yr in range(dob.year, cur.year+1):
    print('##############################################################################')
    print (cl.calendar(yr, 2, 1, 8))

print(bcolors.OKCYAN + 'You have reached the end of program .Thank you' + bcolors.ENDC)
VERBOSE_FUN('THE PROGRAM HAS ENDED')
```
